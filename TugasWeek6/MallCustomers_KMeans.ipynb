```python
import pandas as pd
import numpy as np
```

```python
mall_customers = pd.read_csv('Mall_Customers.csv')
mall_customers.info()
```

```python
mall_customers.rename(columns={
    'Annual Income (k$)' : 'AnnualIncome',
    'Spending Score (1-100)' : 'SpendingScore'
}, inplace=True)
```

```python
variables = mall_customers[['Age', 'AnnualIncome', 'SpendingScore']]
```

```python
from sklearn.cluster import KMeans
```

```python
KMeans_model = KMeans(n_clusters=5, n_init=10)
KMeans_model.fit(variables)
mall_customers['Labels'] = KMeans_model.labels_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.style.use('default')
plt.scatter(mall_customers['AnnualIncome'], mall_customers['SpendingScore'], c=mall_customers['Labels'], s=40)
```
