```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/DATASET/kc_house_data.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
```

```python
# Display basic statistics of the dataset
df.describe()

# Check for missing values
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of house prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, color='skyblue')
plt.title('Histogram of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot: Square footage vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df, color='green')
plt.title('Scatter Plot: Square Footage vs. Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.show()
```

```python
# Box plot: Price distribution based on the number of bedrooms
plt.figure(figsize=(12, 8))
sns.boxplot(x='bedrooms', y='price', data=df, palette='pastel')
plt.title('Box Plot: Price Distribution Based on the Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# List semua fitur
fitur = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement', 'lat', 'long', 'sqft_living15', 'sqft_lot15']

# Buat scatter plot untuk setiap pasang fitur
sns.set(style="ticks")
sns.pairplot(df[fitur])
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
