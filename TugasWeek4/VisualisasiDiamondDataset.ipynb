```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/DATASET/diamonds.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot: Price vs Carat
plt.figure(figsize=(12, 8))
sns.scatterplot(x='carat', y='price', data=df, hue='cut', palette='viridis', size='depth', sizes=(10, 200))
plt.title('Scatter Plot: Price vs Carat with Cut-based Color and Size-based on Depth')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()

# Box plot: Distribution of Price based on Cut
plt.figure(figsize=(12, 8))
sns.boxplot(x='cut', y='price', data=df, palette='Set2')
plt.title('Box Plot: Price Distribution Based on Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```

```python
# Create a Violin plot: Price distribution based on Cut
plt.figure(figsize=(12, 8))
sns.violinplot(x='cut', y='price', data=df, palette='Set2')
plt.title('Violin Plot: Distribution of Price Based on Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```
